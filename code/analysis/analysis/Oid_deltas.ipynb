```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
deltas = np.loadtxt('oid_deltas.csv', skiprows=1, delimiter=',')
labels = ['20160909', '20160911', '20160912']
IPTG = [0, 0.1, 5, 10, 25, 50, 75, 100, 250, 500, 1000, 5000]
```

```python
plt.style.use('ggplot')
for i in range(3):
    plt.semilogx(IPTG, deltas[:, i], 'o-', label=labels[i])
plt.xlabel(r'IPTG ($\mu$M)')
plt.ylabel('Expression')
plt.legend()
```
